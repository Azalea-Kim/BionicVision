```python
%cd {HOME}/Tracking-Anything-with-DEVA

import os
from os import path
from argparse import ArgumentParser

import torch
import numpy as np

from deva.model.network import DEVA
from deva.inference.inference_core import DEVAInferenceCore
from deva.inference.result_utils import ResultSaver
from deva.inference.eval_args import add_common_eval_args, get_model_and_config
from deva.inference.demo_utils import flush_buffer
from deva.ext.ext_eval_args import add_ext_eval_args, add_text_default_args
from deva.ext.grounding_dino import get_grounding_dino_model
from deva.ext.with_text_processor import process_frame_with_text as process_frame

from tqdm import tqdm
import json

torch.autograd.set_grad_enabled(False)

# for id2rgb
np.random.seed(42)

# default parameters
parser = ArgumentParser()
add_common_eval_args(parser)
add_ext_eval_args(parser)
add_text_default_args(parser)

# load model and config
args = parser.parse_args([])
cfg = vars(args)
cfg['enable_long_term'] = True

# Load our checkpoint
deva_model = DEVA(cfg).cuda().eval()
if args.model is not None:
    model_weights = torch.load(args.model)
    deva_model.load_weights(model_weights)
else:
    print('No model loaded.')

gd_model, sam_model = get_grounding_dino_model(cfg, 'cuda')
```

```python
cfg['enable_long_term_count_usage'] = True
cfg['max_num_objects'] = 50
cfg['size'] = 480
cfg['DINO_THRESHOLD'] = 0.35
cfg['amp'] = True
cfg['chunk_size'] = 4
cfg['detection_every'] = 1
cfg['max_missed_detection_count'] = 10
cfg['sam_variant'] = 'original'
cfg['temporal_setting'] = 'online' # semionline usually works better; but online is faster for this demo
cfg['pluralize'] = True
cfg['prompt'] = 'human_arm'
category_len = len(cfg['prompt'].split('.'))

SOURCE_VIDEO_PATH = f"{HOME}/data/rgb_video.mp4"
OUTPUT_VIDEO_PATH = f"{HOME}/data/output.webm"
MASK_DIR = f"{HOME}/data/masks"

%rm -rf MASK_DIR
%rm -rf OUTPUT_VIDEO_PATH
os.makedirs(MASK_DIR, exist_ok=True)
```

```python
from queue import Queue, Empty

class TeeQueue:
    def __init__(self, real_queue: Queue):
        self.real_queue = real_queue      # the queue that the background thread is actually using
        self.user_queue = Queue()         # our own queue to sniff out copies

    def put(self, item, block=True, timeout=None):
        # Put in the real queue (so background thread can consume it)
        self.real_queue.put(item, block=block, timeout=timeout)
        # Also copy it to our user queue
        self.user_queue.put(item)

    def get(self, block=True, timeout=None):
        # The background thread calls queue.get(), so forward that to the real queue
        return self.real_queue.get(block=block, timeout=timeout)

    def qsize(self):
        return self.real_queue.qsize()

    def empty(self):
        return self.real_queue.empty()

    # If you need .task_done() or .join() or other methods, just forward them similarly:
    def task_done(self):
        self.real_queue.task_done()

    def join(self):
        self.real_queue.join()
```

```python
import matplotlib.pyplot as plt
import time
%cd {HOME}/Tracking-Anything-with-DEVA

from deva.ext.with_text_processor import process_frame_with_text as process_frame_text
import tempfile
import cv2

deva = DEVAInferenceCore(deva_model, config=cfg)
deva.next_voting_frame = cfg['num_voting_frames'] - 1
deva.enabled_long_id()

# obtain temporary directory
result_saver = ResultSaver(None, None, dataset='gradio', object_manager=deva.object_manager)
tee_queue = TeeQueue(result_saver.queue)
result_saver.queue = tee_queue
writer_initizied = False

cap = cv2.VideoCapture(SOURCE_VIDEO_PATH)
fps = cap.get(cv2.CAP_PROP_FPS)
ti = 0
# only an estimate
with torch.cuda.amp.autocast(enabled=cfg['amp']):
    with tqdm(total=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))) as pbar:
        while (cap.isOpened()):
            ret, frame = cap.read()
            if ret == True:
                if not writer_initizied:
                    h, w = frame.shape[:2]
                    writer = cv2.VideoWriter(OUTPUT_VIDEO_PATH, cv2.VideoWriter_fourcc(*'vp80'), fps, (w, h))
                    writer_initizied = True
                    result_saver.writer = writer

                process_frame_text(deva, gd_model, sam_model, f'{ti}.png', result_saver, ti, image_np=frame)
                while True:
                  try:
                      args = tee_queue.user_queue.get_nowait()
                  except Empty:
                      break

                  # The background thread will still consume it from the real queue,
                  # but we get a "copy" here
                  # 'args.frame_name' is something like '0.png' or '10.png'
                  frame_idx = int(args.frame_name.split('.')[0])
                  mask_np = np.squeeze(args.mask.cpu().numpy()).astype(np.uint8)

                  # Write it to disk or do whatever you need with it
                  out_path = os.path.join(MASK_DIR, f"frame_{frame_idx:05d}.png")
                  cv2.imwrite(out_path, mask_np)

                  # Mark our user queue item as processed
                  tee_queue.user_queue.task_done()

                ti += 1
                pbar.update(1)
            else:
                break
    flush_buffer(deva, result_saver)
writer.release()
cap.release()
deva.clear_buffer()
```
